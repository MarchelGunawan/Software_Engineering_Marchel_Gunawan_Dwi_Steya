```python
# List 
contoh_list = ["Marchel", 12, 12.0, ["Bambang", "Sugiono"]]
```

```python
print(contoh_list[0])
print(contoh_list[3][0])
print(contoh_list[3][1])
```

```python
for i in contoh_list:
    print(i)
```

```python
# Dictionary
contoh_dict = {
    "name" : "Marchel",
    "Angka" : 12,
    "Tahun" : 2001,
}
```

```python
print(contoh_dict["name"])
print(contoh_dict["Tahun"])
```

```python
# Menambah value ke Dictionary
contoh_dict["name"] = ["Marchel","Bambang"]
contoh_dict
```

```python
print(contoh_dict["name"][1])
```

```python
# Contoh method di dict : .keys()
contoh_dict.keys()
```

```python
# Contoh method di list
arr = [2,3,5,1,4,10]
```

```python
# insert value to list
arr.append(90)
arr
```

```python
# insert value to list but using index
arr.insert(0,100)
arr
```

```python
# sorting list
arr.sort()
arr
```

```python
# Reverse sorting list
arr.sort(reverse = True)
arr
```

```python
# Pop fungsi ny hampir sama kayak di stack
arr.pop()
arr
```

```python
# Remove isi
arr.remove(10)
arr
```

```python
# Tuple
contoh_tuple = ("Marchel",12,12.0,"Bambang")
```

```python
print(contoh_tuple[0])
print(contoh_tuple[1])
```

```python
# Remove duplicate number using set
arr2 = [1,3,1,5,7]
arr2
```

```python
set_angka = set(arr2)
set_angka
```

```python
# Karena set enggak bisa dipanggil secara langsung
arr2 = list(set_angka)
arr2
```

```python
# Looping
for i in range(10):
    print(i)
```

```python
# Looping decrement
for i in range(10,0,-1):
    print(i)
```

```python
# While Looping
while(True):
    print("a")
```

```python
while True:
    a = int(input())
    if a > 20:
        break
    elif a == 10:
        continue
    print("halo")
```

```python
# For else
for i in range(21):
    print(i)
else:
    print("Error")
```

```python
for i in range(21):
    print(i,end=" ")
    if i == 10:
        break
else:
    print("Error")
```
